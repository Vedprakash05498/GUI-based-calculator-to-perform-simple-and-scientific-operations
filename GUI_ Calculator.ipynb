```python
import tkinter as tk

# Functions for performing arithmetic operations
def add(num1, num2):
    return num1 + num2

def subtract(num1, num2):
    return num1 - num2

def multiply(num1, num2):
    return num1 * num2

def divide(num1, num2):
    return num1 / num2

# GUI window setup
root = tk.Tk()
root.title("Simple Calculator")

# Input fields for numbers
num1_field = tk.Entry(root)
num2_field = tk.Entry(root)
num1_field.grid(row=0, column=1)
num2_field.grid(row=1, column=1)

# Labels for input fields
num1_label = tk.Label(root, text="Number 1:")
num2_label = tk.Label(root, text="Number 2:")
num1_label.grid(row=0, column=0)
num2_label.grid(row=1, column=0)

# Dropdown menu for selecting operation
operation = tk.StringVar(root)
operation.set("Select Operation")
operation_menu = tk.OptionMenu(root, operation, "Add", "Subtract", "Multiply", "Divide")
operation_menu.grid(row=2, column=0)

# Output field for result
result = tk.StringVar()
result_label = tk.Label(root, textvariable=result)
result_label.grid(row=3, column=1)

# Function to evaluate selected operation
def evaluate():
    num1 = float(num1_field.get())
    num2 = float(num2_field.get())
    if operation.get() == "Add":
        result.set(add(num1, num2))
    elif operation.get() == "Subtract":
        result.set(subtract(num1, num2))
    elif operation.get() == "Multiply":
        result.set(multiply(num1, num2))
    elif operation.get() == "Divide":
        result.set(divide(num1, num2))

# Button for evaluating selected operation
evaluate_button = tk.Button(root, text="Evaluate", command=evaluate)
evaluate_button.grid(row=2, column=1)

# Start GUI window loop
root.mainloop()
```
